```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('C:/machine learning/WineQt.csv')
df.head()
```

```python
df[df['quality']<=6].head()
```

```python
df[df['quality']>6].head()
```

```python
df.loc[df['quality']<=6,'quality']=0
df.loc[df['quality']>6,'quality']=1
df.head()
```

```python
df[df['quality']==1].head()
```

```python
norm = df.columns[df.columns != 'quality']
scaler = MinMaxScaler()
df[norm] = scaler.fit_transform(df[norm])
df.head()
```
